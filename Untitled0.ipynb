```python
# ================================================================
# CELL 3: DATASET CLASS + DATALOADERS
# ----------------------------------------------------------------
# dataset chhota hai isliye:
# 1. Aggressive augmentation: zyada variety
# 2. Chhota batch size (8): stable training
# 3. CLIP text = folder name directly
# ================================================================

import clip
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T

class SkincareDataset(Dataset):
    def __init__(self, dataframe, preprocess,
                 augment=False, text_map=None):
        self.df         = dataframe.reset_index(drop=True)
        self.preprocess = preprocess
        self.text_map   = text_map or {}

        # ---- Augmentation ----
        # Dataset chhota hai — aggressive augmentation zaroori hai
        if augment:
            self.aug = T.Compose([
                T.RandomHorizontalFlip(p=0.5),
                T.RandomVerticalFlip(p=0.1),
                T.RandomRotation(degrees=20),
                T.ColorJitter(
                    brightness=0.4,
                    contrast=0.4,
                    saturation=0.3,
                    hue=0.08
                ),
                T.RandomResizedCrop(224, scale=(0.75, 1.0)),
                T.RandomGrayscale(p=0.05),
                T.GaussianBlur(kernel_size=3, sigma=(0.1, 1.5)),
            ])
        else:
            self.aug = None

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        # Image load
        try:
            img = Image.open(row['image_path']).convert('RGB')
            if self.aug:
                img = self.aug(img)
            img_tensor = self.preprocess(img)
        except Exception:
            img_tensor = torch.zeros(3, 224, 224)

        # Text — folder name directly use karo
        text_desc  = self.text_map.get(row['label'], row['label'])
        text_token = clip.tokenize([text_desc], truncate=True)[0]

        # Label
        label_idx = LABEL2IDX[row['label']]

        return {
            'image'    : img_tensor,
            'text'     : text_token,
            'label'    : torch.tensor(label_idx, dtype=torch.long),
            'label_str': row['label'],
        }


# Dataset objects
train_ds = SkincareDataset(train_df, clip_preprocess,
                            augment=True,  text_map=SKIN_TEXT)
val_ds   = SkincareDataset(val_df,   clip_preprocess,
                            augment=False, text_map=SKIN_TEXT)
test_ds  = SkincareDataset(test_df,  clip_preprocess,
                            augment=False, text_map=SKIN_TEXT)

# ---- DataLoaders ----
# Batch size = 8 (dataset chhota hai)
train_loader = DataLoader(train_ds, batch_size=8,
                          shuffle=True,  num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds,   batch_size=8,
                          shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=8,
                          shuffle=False, num_workers=2, pin_memory=True)

# Verify
sample = next(iter(train_loader))
print(f"Image shape : {sample['image'].shape}")
print(f"Text shape  : {sample['text'].shape}")
print(f"Label shape : {sample['label'].shape}")
print(f"\nTrain batches : {len(train_loader)}")
print(f"Val   batches : {len(val_loader)}")
print(f"Test  batches : {len(test_loader)}")

# Text prompts check
print(f"\nSample text prompts (CLIP ke liye):")
for label, text in list(SKIN_TEXT.items())[:4]:
    print(f"  '{text}'")

print("✅ Cell 3 complete!")
```

```python
# ================================================================
# CELL 4: ADAPTER LAYER + CLIP MODEL
# ----------------------------------------------------------------
# Adapter kya karta hai?
#
#  CLIP features (512-d)     <- frozen, nahi badlega
#       ↓
#  Down: 512 → 128           <- trainable (compress)
#       ↓
#  GELU activation
#       ↓
#  Up: 128 → 512             <- trainable (expand)
#       ↓
#  Original + alpha × adapted  <- residual connection
#       ↓
#  LayerNorm
#  (512-d adapted features)
#
# CLIP freeze kyun?
# - 400M+ parameters — pura train karna bahut slow + overfit
# - Adapter sirf ~130K parameters — fast + accurate
# - CLIP ki image-text knowledge preserve hoti hai
# ================================================================

class AdapterLayer(nn.Module):
    """
    Bottleneck adapter: CLIP features ko task ke liye adapt karta hai.
    Sirf yeh train hota hai — CLIP nahi.
    """
    def __init__(self, dim=512, bottleneck=128, dropout=0.1):
        super().__init__()
        self.down  = nn.Linear(dim, bottleneck)  # compress
        self.act   = nn.GELU()
        self.drop  = nn.Dropout(dropout)
        self.up    = nn.Linear(bottleneck, dim)  # expand
        self.norm  = nn.LayerNorm(dim)
        # Alpha: model khud seekhta hai kitna adapt karna hai
        self.alpha = nn.Parameter(torch.tensor(0.5))
        self._init()

    def _init(self):
        # Chhote weights se shuru karo — stable training
        nn.init.xavier_uniform_(self.down.weight, gain=0.01)
        nn.init.zeros_(self.down.bias)
        nn.init.xavier_uniform_(self.up.weight, gain=0.01)
        nn.init.zeros_(self.up.bias)

    def forward(self, x):
        residual = x
        out = self.up(self.drop(self.act(self.down(x))))
        # Residual: original features + scaled adapter output
        return self.norm(residual + self.alpha.clamp(0, 1) * out)


class CLIPFewShotAdapter(nn.Module):
    """
    Complete CLIP Few-Shot Model:

    Image → [CLIP Vision FROZEN] → [Image Adapter] ─┐
                                                      ├→ [Fusion] → [Classifier]
    Text  → [CLIP Text  FROZEN] → [Text  Adapter] ─┘

    Few-Shot mode:
    Support set → prototypes (class averages)
    Query image → embedding → cosine similarity → nearest prototype
    """

    def __init__(self, clip_model, num_classes,
                 dim=512, bottleneck=128, dropout=0.15):
        super().__init__()
        self.clip = clip_model

        # CLIP ko FREEZE karo
        for p in self.clip.parameters():
            p.requires_grad = False

        # Trainable adapters
        self.img_adapter = AdapterLayer(dim, bottleneck, dropout)
        self.txt_adapter = AdapterLayer(dim, bottleneck, dropout)

        # Fusion: [img(512) + txt(512)] → 512
        self.fusion = nn.Sequential(
            nn.Linear(dim * 2, dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.LayerNorm(dim)
        )

        # Classifier head
        self.classifier = nn.Sequential(
            nn.Linear(dim, 256),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(256, num_classes)
        )

        # Few-shot ke liye — baad mein fill honge
        self.prototypes   = None
        self.proto_labels = None
        self.proto_matrix = None

        # Parameter count
        trainable = sum(p.numel() for p in self.parameters()
                        if p.requires_grad)
        total     = sum(p.numel() for p in self.parameters())
        print(f"Trainable : {trainable:,} parameters")
        print(f"Total     : {total:,} parameters")
        print(f"Training  : {100*trainable/total:.2f}% of model")

    def encode_image(self, images):
        """CLIP vision (frozen) + image adapter (trainable)"""
        with torch.no_grad():
            feat = self.clip.encode_image(images).float()
        feat = F.normalize(feat, dim=-1)
        return self.img_adapter(feat)

    def encode_text(self, texts):
        """CLIP text (frozen) + text adapter (trainable)"""
        with torch.no_grad():
            feat = self.clip.encode_text(texts).float()
        feat = F.normalize(feat, dim=-1)
        return self.txt_adapter(feat)

    def forward(self, images, texts):
        img_f  = self.encode_image(images)
        txt_f  = self.encode_text(texts)
        fused  = self.fusion(torch.cat([img_f, txt_f], dim=-1))
        logits = self.classifier(fused)
        return logits, img_f, txt_f

    # ── FEW-SHOT: Prototype computation ──────────────────────────
    def build_prototypes(self, support_loader):
        """
        Support set ke har class ka average embedding compute karo.
        Yahi prototype hai — few-shot ka core concept.
        """
        self.eval()
        store = {}
        with torch.no_grad():
            for batch in tqdm(support_loader, desc="Building prototypes"):
                imgs = batch['image'].to(device)
                txts = batch['text'].to(device)
                lbls = batch['label_str']
                img_f = self.encode_image(imgs)
                txt_f = self.encode_text(txts)
                fused = self.fusion(torch.cat([img_f, txt_f], dim=-1))
                for i, lbl in enumerate(lbls):
                    store.setdefault(lbl, []).append(fused[i].cpu())

        self.prototypes = {
            lbl: torch.stack(feats).mean(0)
            for lbl, feats in store.items()
        }
        self.proto_labels = list(self.prototypes.keys())
        self.proto_matrix = torch.stack(
            [self.prototypes[l] for l in self.proto_labels]
        ).to(device)
        print(f"Prototypes: {len(self.prototypes)} classes built")

    def predict_few_shot(self, images, texts):
        """
        Query image → embedding → cosine similarity → nearest prototype.
        """
        self.eval()
        with torch.no_grad():
            img_f = self.encode_image(images)
            txt_f = self.encode_text(texts)
            query = self.fusion(torch.cat([img_f, txt_f], dim=-1))
            q_n   = F.normalize(query, dim=-1)
            p_n   = F.normalize(self.proto_matrix, dim=-1)
            sims  = q_n @ p_n.T                 # (B, num_classes)
            preds = sims.argmax(dim=-1)
            pred_labels = [self.proto_labels[i] for i in preds.cpu()]
        return pred_labels, sims


# ================================================================
# MODEL BANANA
# ================================================================
model = CLIPFewShotAdapter(
    clip_model  = clip_model,
    num_classes = NUM_CLASSES,
    dim         = 512,
    bottleneck  = 128,
    dropout     = 0.15
).to(device)

# Shape test
dummy_img = torch.zeros(2, 3, 224, 224).to(device)
dummy_txt = torch.zeros(2, 77, dtype=torch.long).to(device)
out, _, _ = model(dummy_img, dummy_txt)
print(f"\nOutput shape: {out.shape}")  # (2, num_classes)
print(f"Num classes : {NUM_CLASSES}")
print("✅ Cell 4 complete!")
```

```python
# ================================================================
# CELL 5: COMPLETE TRAINING
# ================================================================

class EarlyStopping:
    def __init__(self, patience=15, save_path='best_clip_adapter.pth'):
        self.patience   = patience
        self.best_loss  = float('inf')
        self.counter    = 0
        self.best_epoch = 0
        self.save_path  = save_path

    def __call__(self, val_loss, model, epoch):
        if val_loss < self.best_loss - 0.001:
            self.best_loss  = val_loss
            self.counter    = 0
            self.best_epoch = epoch
            torch.save({
                'epoch'      : epoch,
                'model_state': model.state_dict(),
                'val_loss'   : val_loss,
                'num_classes': NUM_CLASSES,
                'class_names': CLASS_NAMES,
                'label2idx'  : LABEL2IDX,
                'idx2label'  : IDX2LABEL,
                'skin_text'  : SKIN_TEXT,
            }, self.save_path)
            print(f"      ✅ Saved epoch {epoch} (loss={val_loss:.4f})")
            return False
        else:
            self.counter += 1
            if self.counter >= self.patience:
                print(f"\n   Early stop! Best was epoch {self.best_epoch}")
                return True
            return False


def contrastive_loss(img_f, txt_f, temp=0.07):
    """
    Image aur text ko align karta hai.
    Same class ke image-text pairs ko paas laao.
    """
    img_n = F.normalize(img_f, dim=-1)
    txt_n = F.normalize(txt_f, dim=-1)
    sim   = img_n @ txt_n.T / temp
    tgt   = torch.arange(len(img_n)).to(device)
    return (F.cross_entropy(sim, tgt) + F.cross_entropy(sim.T, tgt)) / 2


def run_epoch(model, loader, optimizer, criterion, training=True):
    model.train() if training else model.eval()
    total_loss, correct, total = 0.0, 0, 0
    ctx = torch.enable_grad() if training else torch.no_grad()

    with ctx:
        for batch in loader:
            imgs = batch['image'].to(device)
            txts = batch['text'].to(device)
            lbls = batch['label'].to(device)

            if training:
                optimizer.zero_grad()

            logits, img_f, txt_f = model(imgs, txts)

            cls_loss = criterion(logits, lbls)
            con_loss = contrastive_loss(img_f, txt_f)
            loss     = cls_loss + 0.1 * con_loss

            if training:
                loss.backward()
                nn.utils.clip_grad_norm_(
                    [p for p in model.parameters() if p.requires_grad], 1.0
                )
                optimizer.step()

            total_loss += loss.item()
            correct    += (logits.argmax(1) == lbls).sum().item()
            total      += lbls.size(0)

    return total_loss / len(loader), correct / total * 100


def train_model(model, epochs=60, lr=5e-4,
                save_path='best_clip_adapter.pth'):

    criterion  = nn.CrossEntropyLoss(label_smoothing=0.1)
    trainable  = [p for p in model.parameters() if p.requires_grad]
    optimizer  = torch.optim.AdamW(trainable, lr=lr, weight_decay=0.01)
    scheduler  = torch.optim.lr_scheduler.CosineAnnealingLR(
                     optimizer, T_max=epochs, eta_min=1e-6)
    early_stop = EarlyStopping(patience=15, save_path=save_path)

    history = {
        'train_loss': [], 'val_loss'  : [],
        'train_acc' : [], 'val_acc'   : [],
        'lr'        : []
    }

    print(f"Training | Epochs:{epochs} | LR:{lr} | Device:{device}")
    print(f"Classes  : {NUM_CLASSES} | Train size: {len(train_df)}")
    print("="*65)

    for epoch in range(1, epochs + 1):

        tr_loss, tr_acc = run_epoch(
            model, train_loader, optimizer, criterion, training=True
        )
        vl_loss, vl_acc = run_epoch(
            model, val_loader, optimizer, criterion, training=False
        )
        scheduler.step()
        cur_lr = scheduler.get_last_lr()[0]

        history['train_loss'].append(tr_loss)
        history['val_loss'].append(vl_loss)
        history['train_acc'].append(tr_acc)
        history['val_acc'].append(vl_acc)
        history['lr'].append(cur_lr)

        if epoch % 5 == 0 or epoch == 1:
            print(f"Ep {epoch:3d} | "
                  f"TrLoss {tr_loss:.4f} TrAcc {tr_acc:.1f}% | "
                  f"VlLoss {vl_loss:.4f} VlAcc {vl_acc:.1f}% | "
                  f"LR {cur_lr:.2e}")

        if early_stop(vl_loss, model, epoch):
            break

    # Best model reload
    ckpt = torch.load(save_path, map_location=device)
    model.load_state_dict(ckpt['model_state'])
    print(f"\nBest model loaded from epoch {early_stop.best_epoch}")
    print(f"Best val_loss : {early_stop.best_loss:.4f}")
    return history


# ================================================================
# TRAINING RUN KARO
# ================================================================
print("Training shuru ho rahi hai...")
history = train_model(
    model     = model,
    epochs    = 60,
    lr        = 5e-4,
    save_path = 'best_clip_adapter.pth'
)
print("✅ Cell 5 complete!")
```

```python
# ================================================================
# CELL 5: FIXED TRAINING — OVERFITTING ROKNE KE LIYE

#
# Solutions applied:
# 1. Zyada dropout (0.15 → 0.4)
# 2. Zyada weight decay (0.01 → 0.1)
# 3. Label smoothing badhaaya (0.1 → 0.2)
# 4. LR kam kiya (5e-4 → 1e-4)
# 5. Contrastive loss weight badhaaya
# 6. Gradient clipping strong kiya
# ================================================================

class EarlyStopping:
    def __init__(self, patience=15, save_path='best_clip_adapter.pth'):
        self.patience   = patience
        self.best_loss  = float('inf')
        self.counter    = 0
        self.best_epoch = 0
        self.save_path  = save_path

    def __call__(self, val_loss, model, epoch):
        if val_loss < self.best_loss - 0.001:
            self.best_loss  = val_loss
            self.counter    = 0
            self.best_epoch = epoch
            torch.save({
                'epoch'      : epoch,
                'model_state': model.state_dict(),
                'val_loss'   : val_loss,
                'num_classes': NUM_CLASSES,
                'class_names': CLASS_NAMES,
                'label2idx'  : LABEL2IDX,
                'idx2label'  : IDX2LABEL,
                'skin_text'  : SKIN_TEXT,
            }, self.save_path)
            print(f"      ✅ Saved epoch {epoch} "
                  f"(val_loss={val_loss:.4f})")
            return False
        else:
            self.counter += 1
            if self.counter >= self.patience:
                print(f"\n   Early stop! Best epoch: {self.best_epoch}")
                return True
            return False


def contrastive_loss(img_f, txt_f, temp=0.07):
    img_n = F.normalize(img_f, dim=-1)
    txt_n = F.normalize(txt_f, dim=-1)
    sim   = img_n @ txt_n.T / temp
    tgt   = torch.arange(len(img_n)).to(device)
    return (F.cross_entropy(sim, tgt) +
            F.cross_entropy(sim.T, tgt)) / 2


def run_epoch(model, loader, optimizer, criterion, training=True):
    model.train() if training else model.eval()
    total_loss, correct, total = 0.0, 0, 0
    ctx = torch.enable_grad() if training else torch.no_grad()

    with ctx:
        for batch in loader:
            imgs = batch['image'].to(device)
            txts = batch['text'].to(device)
            lbls = batch['label'].to(device)

            if training:
                optimizer.zero_grad()

            logits, img_f, txt_f = model(imgs, txts)

            cls_loss = criterion(logits, lbls)
            con_loss = contrastive_loss(img_f, txt_f)

            # Contrastive loss weight badhaaya — 0.1 → 0.3
            # Yeh image-text alignment strong karta hai
            loss = cls_loss + 0.3 * con_loss

            if training:
                loss.backward()
                # Strong gradient clipping
                nn.utils.clip_grad_norm_(
                    [p for p in model.parameters()
                     if p.requires_grad], 0.5
                )
                optimizer.step()

            total_loss += loss.item()
            correct    += (logits.argmax(1) == lbls).sum().item()
            total      += lbls.size(0)

    return total_loss / len(loader), correct / total * 100


def train_model(model, epochs=60, lr=1e-4,
                save_path='best_clip_adapter.pth'):

    # Label smoothing 0.2 — model ko overconfident hone se rokta hai
    criterion = nn.CrossEntropyLoss(label_smoothing=0.2)

    trainable = [p for p in model.parameters() if p.requires_grad]

    # Weight decay 0.1 — strong regularization
    optimizer = torch.optim.AdamW(
        trainable, lr=lr, weight_decay=0.1
    )

    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
        optimizer, T_max=epochs, eta_min=1e-7
    )

    early_stop = EarlyStopping(patience=15, save_path=save_path)

    history = {
        'train_loss': [], 'val_loss': [],
        'train_acc' : [], 'val_acc' : [],
        'lr'        : []
    }

    print(f"Training | Epochs:{epochs} | LR:{lr} | Device:{device}")
    print(f"Classes  : {NUM_CLASSES} | Train: {len(train_df)} images")
    print(f"Regularization: dropout=0.4, weight_decay=0.1, "
          f"label_smooth=0.2")
    print("="*65)
    print(f"{'Epoch':>6} | {'TrLoss':>7} {'TrAcc':>7} | "
          f"{'VlLoss':>7} {'VlAcc':>7} | {'LR':>9}")
    print("-"*65)

    for epoch in range(1, epochs + 1):

        tr_loss, tr_acc = run_epoch(
            model, train_loader, optimizer, criterion, True
        )
        vl_loss, vl_acc = run_epoch(
            model, val_loader, optimizer, criterion, False
        )
        scheduler.step()
        cur_lr = scheduler.get_last_lr()[0]

        history['train_loss'].append(tr_loss)
        history['val_loss'].append(vl_loss)
        history['train_acc'].append(tr_acc)
        history['val_acc'].append(vl_acc)
        history['lr'].append(cur_lr)

        # Har epoch print karo — chhota dataset hai
        if epoch % 5 == 0 or epoch == 1:
            print(f"Ep {epoch:3d} | "
                  f"{tr_loss:7.4f} {tr_acc:6.1f}% | "
                  f"{vl_loss:7.4f} {vl_acc:6.1f}% | "
                  f"{cur_lr:.2e}")

        if early_stop(vl_loss, model, epoch):
            break

    # Best model reload
    ckpt = torch.load(save_path, map_location=device)
    model.load_state_dict(ckpt['model_state'])
    print(f"\nBest model: epoch {early_stop.best_epoch} | "
          f"val_loss={early_stop.best_loss:.4f}")
    return history


# ================================================================
# MODEL REBUILD KARO — Zyada Dropout Ke Saath
# ================================================================
# Pehle model delete karo
del model
torch.cuda.empty_cache()

# Naya model — dropout 0.4 (pehle 0.15 tha)
model = CLIPFewShotAdapter(
    clip_model  = clip_model,
    num_classes = NUM_CLASSES,
    dim         = 512,
    bottleneck  = 128,
    dropout     = 0.4        # ← badha diya 0.15 se 0.4
).to(device)

print("\nTraining shuru ho rahi hai...")
history = train_model(
    model     = model,
    epochs    = 60,
    lr        = 1e-4,        # ← kam kiya 5e-4 se
    save_path = 'best_clip_adapter.pth'
)
print("✅ Cell 5 complete!")
```

```python
# ================================================================
# CELL 7: TEST PREDICTIONS
# Standard classifier + Few-Shot predictions dono collect karo
# ================================================================

def collect_predictions_clip(model, loader):
    """CLIP Adapter — image + text dono chahiye"""
    model.eval()
    all_true, all_pred, all_prob = [], [], []
    with torch.no_grad():
        for batch in tqdm(loader, desc="Collecting", leave=False):
            imgs  = batch['image'].to(device)
            txts  = batch['text'].to(device)
            lbls  = batch['label']
            logits, _, _ = model(imgs, txts)
            probs = F.softmax(logits, -1).cpu().numpy()
            preds = logits.argmax(1).cpu().numpy()
            all_true.extend(lbls.numpy())
            all_pred.extend(preds)
            all_prob.extend(probs)
    return np.array(all_true), np.array(all_pred), np.array(all_prob)


# ---- Standard classifier predictions ----
print("Standard predictions...")
std_true, std_pred, std_prob = collect_predictions_clip(model, test_loader)
print(f"Standard Accuracy: {accuracy_score(std_true, std_pred)*100:.2f}%")

# ---- Few-Shot predictions ----
print("\nBuilding few-shot prototypes from training data...")
model.build_prototypes(train_loader)

fs_pred_list, fs_prob_list, fs_true_list = [], [], []
model.eval()
for batch in tqdm(test_loader, desc="Few-shot predict"):
    imgs = batch['image'].to(device)
    txts = batch['text'].to(device)
    lbls = batch['label']
    pred_labels, sims = model.predict_few_shot(imgs, txts)
    probs = F.softmax(sims, dim=-1).cpu().numpy()
    preds = [model.proto_labels.index(l) for l in pred_labels]
    fs_pred_list.extend(preds)
    fs_prob_list.extend(probs)
    fs_true_list.extend(lbls.numpy())

fs_true = np.array(fs_true_list)
fs_pred = np.array(fs_pred_list)
fs_prob = np.array(fs_prob_list)
print(f"Few-Shot Accuracy: {accuracy_score(fs_true, fs_pred)*100:.2f}%")
print("✅ Cell 7 complete!")
```

```python
# ================================================================
# CELL 9: PRODUCT DATABASE + RECOMMENDATION FUNCTION
# ================================================================

products_db = pd.DataFrame([
    {"name":"Oil-Free Moisturizer","brand":"CeraVe",
     "skin_type":"oily","concern":"acne",
     "ingredients":"Niacinamide, Salicylic Acid",
     "price":2500,"rating":4.5,"why":"Controls oil, clears pores"},
    {"name":"BHA Exfoliant Toner","brand":"Paula's Choice",
     "skin_type":"oily","concern":"pores",
     "ingredients":"BHA 2%, Niacinamide",
     "price":3200,"rating":4.7,"why":"Minimizes pores, controls oil"},
    {"name":"Hydrating Night Cream","brand":"Neutrogena",
     "skin_type":"dry","concern":"wrinkles",
     "ingredients":"Hyaluronic Acid, Retinol, Ceramides",
     "price":1800,"rating":4.3,"why":"Deep hydration, anti-aging"},
    {"name":"Vitamin C Brightening Serum","brand":"The Ordinary",
     "skin_type":"dry","concern":"dark_spots",
     "ingredients":"15% Vitamin C, Ferulic Acid",
     "price":1200,"rating":4.4,"why":"Fades dark spots, brightens"},
    {"name":"Cicaplast Soothing Cream","brand":"La Roche-Posay",
     "skin_type":"sensitive","concern":"redness",
     "ingredients":"Thermal Water, Ceramides, Panthenol",
     "price":2800,"rating":4.6,"why":"Calms redness, repairs barrier"},
    {"name":"Clear Complexion Moisturizer","brand":"Aveeno",
     "skin_type":"sensitive","concern":"acne",
     "ingredients":"Oat Extract, Zinc, Salicylic Acid",
     "price":1500,"rating":4.2,"why":"Gentle, clears breakouts"},
    {"name":"Pore Refining Concentrate","brand":"Clinique",
     "skin_type":"combination","concern":"pores",
     "ingredients":"Niacinamide, Salicylic Acid, Zinc",
     "price":3500,"rating":4.4,"why":"Balances T-zone, refines pores"},
    {"name":"Even Skin Tone Serum","brand":"Murad",
     "skin_type":"combination","concern":"dark_spots",
     "ingredients":"Tranexamic Acid, Vitamin C",
     "price":4200,"rating":4.5,"why":"Evens tone on combo skin"},
    {"name":"UV Daily SPF 46","brand":"EltaMD",
     "skin_type":"normal","concern":"dark_spots",
     "ingredients":"Zinc Oxide, Niacinamide, SPF 46",
     "price":4000,"rating":4.8,"why":"Sun protection + brightening"},
    {"name":"Retinol Correxion Cream","brand":"RoC",
     "skin_type":"normal","concern":"wrinkles",
     "ingredients":"Retinol, Peptides, Vitamin E",
     "price":2200,"rating":4.3,"why":"Proven anti-aging formula"},
    {"name":"Anti-Redness Serum","brand":"Avene",
     "skin_type":"sensitive","concern":"redness",
     "ingredients":"Diacerhein, Ceramides",
     "price":3000,"rating":4.4,"why":"Reduces visible redness"},
    {"name":"Salicylic Cleanser","brand":"Neutrogena",
     "skin_type":"oily","concern":"acne",
     "ingredients":"Salicylic Acid 2%, Glycerin",
     "price":900,"rating":4.3,"why":"Deep cleans oily acne skin"},
])


def predict_and_recommend(image_input, text_query=None, top_k=3):
    """
    Pipeline:
    Image → CLIP → Adapter → prediction → product match → recommend

    image_input: file path (str) ya PIL Image
    text_query: optional user ka description
    top_k: kitne products recommend karein
    """
    # Image prepare karo
    if isinstance(image_input, str):
        pil_img = Image.open(image_input).convert('RGB')
    else:
        pil_img = image_input.convert('RGB')

    img_tensor = clip_preprocess(pil_img).unsqueeze(0).to(device)

    # Text
    if text_query:
        text_desc = text_query
    else:
        text_desc = "a photo of skin showing skin condition"
    text_token = clip.tokenize([text_desc], truncate=True).to(device)

    # Predict
    model.eval()
    with torch.no_grad():
        logits, _, _ = model(img_tensor, text_token)
        probs = F.softmax(logits, dim=-1)
        conf, idx = probs.max(-1)

    pred_label = IDX2LABEL[idx.item()]
    parts      = pred_label.split('_', 1)
    skin_type  = parts[0]
    concern    = parts[1] if len(parts) > 1 else 'general'

    # Top-3 predictions
    top3_prob, top3_idx = probs[0].topk(min(3, NUM_CLASSES))
    top3 = [(IDX2LABEL[i.item()], round(p.item()*100, 1))
            for i, p in zip(top3_idx, top3_prob)]

    # Product match
    match = products_db[
        (products_db['skin_type'] == skin_type) &
        (products_db['concern']   == concern)
    ].sort_values('rating', ascending=False)

    if len(match) == 0:
        match = products_db[
            products_db['skin_type'] == skin_type
        ].sort_values('rating', ascending=False)

    recs = match.head(top_k).to_dict('records')

    return {
        'skin_type' : skin_type,
        'concern'   : concern,
        'confidence': round(conf.item()*100, 1),
        'top3'      : top3,
        'products'  : recs,
    }


# Quick test
sample_img = test_df.iloc[0]['image_path']
result     = predict_and_recommend(sample_img, top_k=3)
print(f"Skin Type  : {result['skin_type'].upper()}")
print(f"Concern    : {result['concern'].upper()}")
print(f"Confidence : {result['confidence']}%")
for i, p in enumerate(result['products'], 1):
    print(f"  {i}. {p['name']} ({p['brand']}) Rs.{p['price']}")
print("✅ Cell 9 complete!")
```
