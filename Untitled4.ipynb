```python
# ================================================================
# CELL 2: APNA DATASET LOAD KARO
#
# Aapke dataset ka format kuch bhi ho sakta hai:
# Option A: CSV file + image folder  (most common)
# Option B: ImageFolder format (subfolders = classes)
# Option C: Google Drive se load karna
#
# Neeche aapko sirf apna format batana hai
# ================================================================

# ------- GOOGLE DRIVE MOUNT (agar Drive mein dataset hai) -------
from google.colab import drive
drive.mount('/content/drive')

# ================================================================
# OPTION A: CSV + Image Folder format
# CSV mein columns: image_path, skin_type, concern, label
# ================================================================

def load_dataset_from_csv(csv_path, image_base_dir):
    """
    Aapki CSV file se dataset load karta hai.

    CSV format expected:
    image_path       | skin_type   | concern     | label
    img001.jpg       | oily        | acne        | oily_acne
    img002.jpg       | dry         | wrinkles    | dry_wrinkles

    Agar aapki CSV alag format mein hai to column names adjust karein
    """
    df = pd.read_csv(csv_path)

    # ---- YAHAN APNE COLUMN NAMES DALEIN ----
    # Agar aapke columns alag hain to rename karein:
    # df = df.rename(columns={
    #     'your_image_col': 'image_path',
    #     'your_skintype_col': 'skin_type',
    #     'your_concern_col': 'concern'
    # })

    # Agar 'label' column nahi hai to banana padega
    if 'label' not in df.columns:
        df['label'] = df['skin_type'] + '_' + df['concern']

    # Agar 'text_description' nahi hai to auto-generate karo
    if 'text_description' not in df.columns:
        df['text_description'] = df.apply(
            lambda r: f"skin image showing {r['skin_type']} skin with {r['concern']}", axis=1
        )

    # Image paths complete banao
    df['image_path'] = df['image_path'].apply(
        lambda p: os.path.join(image_base_dir, p) if not os.path.isabs(p) else p
    )

    # Missing images remove karo
    df = df[df['image_path'].apply(os.path.exists)].reset_index(drop=True)

    print(f"Loaded: {len(df)} samples")
    print(f"Classes: {df['label'].nunique()}")
    print(f"Distribution:\n{df['label'].value_counts()}")

    return df

# ================================================================
# OPTION B: ImageFolder format (subfolders = class names)
# data/
#   oily_acne/    <- yeh class name hai
#     img001.jpg
#     img002.jpg
#   dry_wrinkles/
#     img003.jpg
# ================================================================

def load_dataset_from_folders(root_dir):
    """
    Folder structure se dataset banata hai.
    Folder ka naam = label
    """
    records = []
    for class_name in os.listdir(root_dir):
        class_dir = os.path.join(root_dir, class_name)
        if not os.path.isdir(class_dir):
            continue

        # class_name format: "oily_acne" ya "dry_wrinkles" etc.
        parts = class_name.split('_', 1)
        skin_type = parts[0] if len(parts) > 0 else class_name
        concern   = parts[1] if len(parts) > 1 else 'general'

        for fname in os.listdir(class_dir):
            if fname.lower().endswith(('.jpg','.jpeg','.png','.bmp')):
                records.append({
                    'image_path': os.path.join(class_dir, fname),
                    'skin_type': skin_type,
                    'concern': concern,
                    'label': class_name,
                    'text_description': f"skin image showing {skin_type} skin with {concern}"
                })

    df = pd.DataFrame(records)
    print(f"Loaded: {len(df)} samples")
    print(f"Classes: {df['label'].nunique()}")
    print(f"Distribution:\n{df['label'].value_counts()}")
    return df


# ================================================================
# APNA PATH YAHAN DALEIN:
# ================================================================

# Option A use kar rahe hain to:
CSV_PATH  = "/content/drive/MyDrive/skincare_dataset/labels.csv"
IMAGE_DIR = "/content/drive/MyDrive/skincare_dataset/images/"

# Option B use kar rahe hain to:
DATASET_ROOT = "/content/drive/MyDrive/skincare_dataset/"

# ---- Ek option uncomment karein ----
# df_full = load_dataset_from_csv(CSV_PATH, IMAGE_DIR)
# df_full = load_dataset_from_folders(DATASET_ROOT)

# ================================================================
# AGAR ABHI SIRF TEST KARNA HAI (demo data se):
# ================================================================
# Demo data banana (baad mein apna data load karo)
from PIL import ImageDraw
import random

os.makedirs("data/images", exist_ok=True)
SKIN_TYPES   = ["oily","dry","combination","sensitive","normal"]
SKIN_CONCERNS= ["acne","wrinkles","dark_spots","redness","pores"]
records = []

for st in SKIN_TYPES:
    for sc in SKIN_CONCERNS:
        for i in range(10):   # 10 per class
            fname = f"data/images/{st}_{sc}_{i}.jpg"
            color_map = {"oily":(220,185,140),"dry":(200,158,120),
                         "combination":(210,170,130),"sensitive":(230,148,140),"normal":(210,170,135)}
            img = Image.new('RGB',(224,224), color_map[st])
            draw = ImageDraw.Draw(img)
            if sc=="acne":
                for _ in range(random.randint(5,15)):
                    x,y=random.randint(20,200),random.randint(20,200)
                    draw.ellipse([x,y,x+8,y+8],fill=(180,70,70))
            img.save(fname)
            records.append({'image_path':fname,'skin_type':st,'concern':sc,
                            'label':f"{st}_{sc}",
                            'text_description':f"skin image showing {st} skin with {sc}"})

df_full = pd.DataFrame(records)
print(f"Demo dataset: {len(df_full)} samples, {df_full['label'].nunique()} classes")

# ================================================================
# TRAIN / TEST SPLIT
# ================================================================
# Stratified split: har class ka same ratio train/test mein
train_df, test_df = train_test_split(
    df_full, test_size=0.2,
    stratify=df_full['label'], random_state=42
)
train_df = train_df.reset_index(drop=True)
test_df  = test_df.reset_index(drop=True)

print(f"\nTrain: {len(train_df)} | Test: {len(test_df)}")
```

```python
# ================================================================
# CELL 3: PYTORCH DATASET CLASS
#
# Yeh class batati hai PyTorch ko:
# - Image kaise load karein
# - Text kaise tokenize karein
# - Batch mein kya return karein
# ================================================================

class SkincareDataset(Dataset):
    def __init__(self, dataframe, preprocess, augment=False):
        """
        dataframe  : pandas DataFrame (image_path, skin_type, concern, label, text_description)
        preprocess : CLIP ka preprocess function
        augment    : training mein data augmentation on/off
        """
        self.df = dataframe.reset_index(drop=True)
        self.preprocess = preprocess
        self.augment = augment

        # Label <-> index mapping
        all_labels = sorted(self.df['label'].unique())
        self.label2idx = {l:i for i,l in enumerate(all_labels)}
        self.idx2label = {i:l for l,i in self.label2idx.items()}
        self.num_classes = len(all_labels)

        # Augmentation (training mein zyada variety ke liye)
        if augment:
            from torchvision import transforms
            self.aug = transforms.Compose([
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.RandomRotation(15),
                transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.1),
                transforms.RandomResizedCrop(224, scale=(0.85, 1.0)),
            ])
        else:
            self.aug = None

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        # Image load karo
        try:
            img = Image.open(row['image_path']).convert('RGB')
            if self.aug:
                img = self.aug(img)
            img_tensor = self.preprocess(img)       # (3, 224, 224)
        except:
            img_tensor = torch.zeros(3, 224, 224)

        # Text tokenize karo
        text_token = clip.tokenize([row['text_description']], truncate=True)[0]  # (77,)

        label_idx = self.label2idx[row['label']]

        return {
            'image'    : img_tensor,
            'text'     : text_token,
            'label'    : torch.tensor(label_idx, dtype=torch.long),
            'label_str': row['label'],
            'skin_type': row['skin_type'],
            'concern'  : row['concern'],
        }


# Dataset aur DataLoader banana
train_ds = SkincareDataset(train_df, preprocess, augment=True)
test_ds  = SkincareDataset(test_df,  preprocess, augment=False)

train_loader = DataLoader(train_ds, batch_size=32, shuffle=True,  num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=32, shuffle=False, num_workers=2, pin_memory=True)

NUM_CLASSES = train_ds.num_classes
CLASS_NAMES = [train_ds.idx2label[i] for i in range(NUM_CLASSES)]

print(f"Train batches : {len(train_loader)}")
print(f"Test  batches : {len(test_loader)}")
print(f"Num classes   : {NUM_CLASSES}")
print(f"Classes       : {CLASS_NAMES}")
```

```python
# ================================================================
# CELL 4: ADAPTER LAYER + FULL MODEL
#
# Adapter layer ka concept:
# ---------------------------------------------------
#  CLIP features (512-d)
#       |
#   [Down: 512->128]   <-- compress karo (learn important features)
#       |
#    [ReLU]
#       |
#   [Up: 128->512]     <-- wapas original size
#       |
#   + residual (original features bhi add karo)
#       |
#  [LayerNorm]
#  (512-d adapted features)
# ---------------------------------------------------
# Sirf yeh ~131K parameters train hote hain, CLIP ke 400M nahi!
# ================================================================

class AdapterLayer(nn.Module):
    def __init__(self, dim=512, bottleneck=128, dropout=0.1):
        super().__init__()
        self.down   = nn.Linear(dim, bottleneck)
        self.act    = nn.GELU()          # GELU smoother hai ReLU se
        self.drop   = nn.Dropout(dropout)
        self.up     = nn.Linear(bottleneck, dim)
        self.norm   = nn.LayerNorm(dim)
        # alpha: model khud seekhta hai adapter ka kitna contribution ho
        self.alpha  = nn.Parameter(torch.tensor(0.5))
        self._init_weights()

    def _init_weights(self):
        # Chhote weights se shuru karo — stable training ke liye
        nn.init.xavier_uniform_(self.down.weight, gain=0.01)
        nn.init.zeros_(self.down.bias)
        nn.init.xavier_uniform_(self.up.weight, gain=0.01)
        nn.init.zeros_(self.up.bias)

    def forward(self, x):
        residual = x
        out = self.down(x)
        out = self.act(out)
        out = self.drop(out)
        out = self.up(out)
        # Residual connection: original + scaled adapter output
        out = self.norm(residual + self.alpha.clamp(0,1) * out)
        return out


class CLIPAdapterModel(nn.Module):
    """
    Complete Model Architecture:

    Image ──> [CLIP Vision Encoder FROZEN] ──> [Image Adapter TRAIN] ──┐
                                                                        ├──> [Fusion] ──> [Classifier] ──> predictions
    Text  ──> [CLIP Text  Encoder FROZEN]  ──> [Text  Adapter TRAIN] ──┘
    """

    def __init__(self, clip_model, num_classes,
                 dim=512, bottleneck=128, dropout=0.1):
        super().__init__()
        self.clip = clip_model

        # CLIP FREEZE karo — bilkul touch nahi hoga
        for p in self.clip.parameters():
            p.requires_grad = False

        # Trainable adapters
        self.img_adapter  = AdapterLayer(dim, bottleneck, dropout)
        self.txt_adapter  = AdapterLayer(dim, bottleneck, dropout)

        # Fusion: image + text concatenate (1024-d) -> 512-d
        self.fusion = nn.Sequential(
            nn.Linear(dim*2, dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.LayerNorm(dim)
        )

        # Classifier head
        self.classifier = nn.Sequential(
            nn.Linear(dim, 256),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(256, num_classes)
        )

        # Trainable param count
        train_p = sum(p.numel() for p in self.parameters() if p.requires_grad)
        total_p = sum(p.numel() for p in self.parameters())
        print(f"Trainable: {train_p:,} / Total: {total_p:,} ({100*train_p/total_p:.2f}%)")

    def encode_image(self, images):
        with torch.no_grad():
            feats = self.clip.encode_image(images).float()
        feats = F.normalize(feats, dim=-1)
        return self.img_adapter(feats)

    def encode_text(self, texts):
        with torch.no_grad():
            feats = self.clip.encode_text(texts).float()
        feats = F.normalize(feats, dim=-1)
        return self.txt_adapter(feats)

    def forward(self, images, texts):
        img_f = self.encode_image(images)   # (B, 512)
        txt_f = self.encode_text(texts)     # (B, 512)
        fused = self.fusion(torch.cat([img_f, txt_f], dim=-1))  # (B, 512)
        logits = self.classifier(fused)     # (B, num_classes)
        return logits, img_f, txt_f

    # -------------------------------------------------------
    # FEW-SHOT: Prototype computation
    # Prototype = har class ka average embedding
    # -------------------------------------------------------
    def build_prototypes(self, support_loader):
        """
        Few-shot ke liye prototypes banao.
        Support set = aapke training examples (N per class).
        Prototype = unka average embedding.
        """
        self.eval()
        store = {}
        with torch.no_grad():
            for batch in tqdm(support_loader, desc="Building prototypes"):
                imgs = batch['image'].to(device)
                txts = batch['text'].to(device)
                lbls = batch['label_str']

                img_f = self.encode_image(imgs)
                txt_f = self.encode_text(txts)
                fused = self.fusion(torch.cat([img_f, txt_f], dim=-1))

                for i, lbl in enumerate(lbls):
                    store.setdefault(lbl, []).append(fused[i].cpu())

        self.prototypes = {
            lbl: torch.stack(feats).mean(0)
            for lbl, feats in store.items()
        }
        self.proto_labels = list(self.prototypes.keys())
        self.proto_matrix = torch.stack(
            [self.prototypes[l] for l in self.proto_labels]
        ).to(device)  # (C, 512)

        print(f"Prototypes built: {len(self.prototypes)} classes")
        return self.prototypes

    def predict_few_shot(self, images, texts):
        """
        Query embedding banao, prototype se cosine similarity compute karo.
        Sabse paas wali class = prediction.
        """
        self.eval()
        with torch.no_grad():
            img_f = self.encode_image(images)
            txt_f = self.encode_text(texts)
            query = self.fusion(torch.cat([img_f, txt_f], dim=-1))

            q_norm = F.normalize(query, dim=-1)
            p_norm = F.normalize(self.proto_matrix, dim=-1)
            sims   = q_norm @ p_norm.T          # (B, C)

            pred_idx = sims.argmax(dim=-1)
            pred_labels = [self.proto_labels[i] for i in pred_idx.cpu()]

        return pred_labels, sims


# Model banana
model = CLIPAdapterModel(clip_model, NUM_CLASSES).to(device)
```

```python
# ================================================================
# CELL 5: TRAINING LOOP
#
# Kya ho raha hai:
# 1. Batch aata hai (images + texts + labels)
# 2. Model forward pass karta hai
# 3. CrossEntropy loss compute hoti hai (galat prediction = zyada loss)
# 4. Backprop: adapter weights adjust hote hain
# 5. Scheduler learning rate smoothly kam karta hai
# ================================================================

def train(model, loader, epochs=30, lr=1e-3, warmup_epochs=3):
    criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
    # Label smoothing: model ko overconfident hone se rokta hai

    trainable = [p for p in model.parameters() if p.requires_grad]
    optimizer = torch.optim.AdamW(trainable, lr=lr, weight_decay=0.01)

    # Cosine scheduler: LR smoothly zero ki taraf jaata hai
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
        optimizer, T_max=epochs, eta_min=1e-6
    )

    history = {'loss': [], 'acc': [], 'lr': []}

    for epoch in range(epochs):
        model.train()
        tot_loss, correct, total = 0.0, 0, 0

        for batch in tqdm(loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False):
            imgs = batch['image'].to(device)
            txts = batch['text'].to(device)
            lbls = batch['label'].to(device)

            optimizer.zero_grad()

            logits, img_f, txt_f = model(imgs, txts)

            # Main classification loss
            cls_loss = criterion(logits, lbls)

            # Contrastive loss (image-text alignment maintain karo)
            # Same label wale pairs similar hone chahiye
            img_n = F.normalize(img_f, dim=-1)
            txt_n = F.normalize(txt_f, dim=-1)
            sim_matrix = img_n @ txt_n.T / 0.07
            targets = torch.arange(len(lbls)).to(device)
            con_loss = (F.cross_entropy(sim_matrix, targets) +
                        F.cross_entropy(sim_matrix.T, targets)) / 2

            loss = cls_loss + 0.1 * con_loss
            loss.backward()

            # Gradient clipping: training stable rakhta hai
            torch.nn.utils.clip_grad_norm_(trainable, 1.0)
            optimizer.step()

            tot_loss += loss.item()
            correct  += (logits.argmax(1) == lbls).sum().item()
            total    += lbls.size(0)

        scheduler.step()

        avg_loss = tot_loss / len(loader)
        avg_acc  = correct / total * 100
        cur_lr   = scheduler.get_last_lr()[0]

        history['loss'].append(avg_loss)
        history['acc'].append(avg_acc)
        history['lr'].append(cur_lr)

        if (epoch+1) % 5 == 0:
            print(f"Epoch {epoch+1:3d} | Loss {avg_loss:.4f} | "
                  f"Acc {avg_acc:.2f}% | LR {cur_lr:.2e}")

    return history


print("Training shuru ho rahi hai...")
history = train(model, train_loader, epochs=30, lr=1e-3)

# Training curves plot karo
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(history['loss'], color='#E24B4A', lw=2)
ax1.set(title='Training Loss', xlabel='Epoch', ylabel='Loss')
ax1.grid(alpha=0.3)
ax2.plot(history['acc'], color='#1D9E75', lw=2)
ax2.set(title='Training Accuracy', xlabel='Epoch', ylabel='Accuracy (%)')
ax2.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('training_curves.png', dpi=150)
plt.show()
```

```python
# ================================================================
# CELL 6: BASELINE / COMPARISON MODELS
# Inse aapka CLIP+Adapter compare hoga
#
# Models:
# 1. ResNet-50
# 2. EfficientNet-B0
# 3. ViT-B/16
# 4. MobileNetV3
# 5. SVM + HOG features
# 6. Zero-Shot CLIP (bina adapter ke)
# ================================================================

import torchvision.models as tvm
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# ---- Helper: kisi bhi PyTorch model ko train karo ----
def train_baseline(model_pt, loader, epochs=10, lr=1e-3, name="Model"):
    """Generic training loop for baseline classifiers"""
    model_pt = model_pt.to(device)
    crit = nn.CrossEntropyLoss()
    opt  = torch.optim.AdamW(model_pt.parameters(), lr=lr, weight_decay=0.01)
    sch  = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=epochs)

    print(f"\nTraining {name}...")
    for epoch in range(epochs):
        model_pt.train()
        correct, total = 0, 0
        for batch in tqdm(loader, desc=f"  {name} Ep{epoch+1}", leave=False):
            imgs = batch['image'].to(device)
            lbls = batch['label'].to(device)
            opt.zero_grad()
            out  = model_pt(imgs)
            loss = crit(out, lbls)
            loss.backward()
            opt.step()
            correct += (out.argmax(1)==lbls).sum().item()
            total   += lbls.size(0)
        sch.step()
    print(f"  {name} trained. Final train acc: {correct/total*100:.1f}%")
    return model_pt

# ---- Helper: predictions aur probabilities collect karo ----
def get_predictions(model_pt, loader, use_softmax=True):
    """Model se all predictions + probabilities collect karo"""
    model_pt.eval()
    all_probs, all_preds, all_true = [], [], []
    with torch.no_grad():
        for batch in loader:
            imgs = batch['image'].to(device)
            lbls = batch['label']
            out  = model_pt(imgs)
            probs = F.softmax(out, dim=-1).cpu().numpy()
            preds = out.argmax(1).cpu().numpy()
            all_probs.extend(probs)
            all_preds.extend(preds)
            all_true.extend(lbls.numpy())
    return np.array(all_true), np.array(all_preds), np.array(all_probs)


# ================================================================
# MODEL 1: ResNet-50
# ================================================================
resnet = tvm.resnet50(weights=tvm.ResNet50_Weights.IMAGENET1K_V1)
resnet.fc = nn.Linear(resnet.fc.in_features, NUM_CLASSES)
resnet = train_baseline(resnet, train_loader, epochs=10, name="ResNet-50")


# ================================================================
# MODEL 2: EfficientNet-B0
# ================================================================
effnet = tvm.efficientnet_b0(weights=tvm.EfficientNet_B0_Weights.IMAGENET1K_V1)
effnet.classifier[1] = nn.Linear(effnet.classifier[1].in_features, NUM_CLASSES)
effnet = train_baseline(effnet, train_loader, epochs=10, name="EfficientNet-B0")


# ================================================================
# MODEL 3: ViT-B/16
# ================================================================
vit = tvm.vit_b_16(weights=tvm.ViT_B_16_Weights.IMAGENET1K_V1)
vit.heads.head = nn.Linear(vit.heads.head.in_features, NUM_CLASSES)
vit = train_baseline(vit, train_loader, epochs=10, lr=5e-5, name="ViT-B/16")


# ================================================================
# MODEL 4: MobileNetV3-Large
# ================================================================
mobilenet = tvm.mobilenet_v3_large(
    weights=tvm.MobileNet_V3_Large_Weights.IMAGENET1K_V1)
mobilenet.classifier[3] = nn.Linear(
    mobilenet.classifier[3].in_features, NUM_CLASSES)
mobilenet = train_baseline(mobilenet, train_loader, epochs=10, name="MobileNetV3")


# ================================================================
# MODEL 5: SVM + CLIP features (HOG-style baseline)
# CLIP se features nikalo, SVM se classify karo
# ================================================================
print("\nTraining SVM + CLIP features...")

def extract_clip_features(loader):
    """CLIP se image features extract karo (frozen)"""
    all_feats, all_labels = [], []
    clip_model.eval()
    with torch.no_grad():
        for batch in tqdm(loader, desc="  Extracting CLIP features"):
            imgs = batch['image'].to(device)
            lbls = batch['label']
            feats = clip_model.encode_image(imgs).float().cpu().numpy()
            all_feats.append(feats)
            all_labels.append(lbls.numpy())
    return np.vstack(all_feats), np.concatenate(all_labels)

train_feats, train_lbls_svm = extract_clip_features(train_loader)
test_feats,  test_lbls_svm  = extract_clip_features(test_loader)

svm_clf = make_pipeline(
    StandardScaler(),
    SVC(kernel='rbf', C=10.0, gamma='scale',
        probability=True, random_state=42)
)
svm_clf.fit(train_feats, train_lbls_svm)
print("  SVM trained!")
svm_preds = svm_clf.predict(test_feats)
svm_probs = svm_clf.predict_proba(test_feats)


# ================================================================
# MODEL 6: Zero-Shot CLIP (bina kisi training ke)
# Text prompts se directly classify karo
# ================================================================
print("\nZero-Shot CLIP evaluation...")

def zero_shot_clip_predict(loader, class_names, clip_model, device):
    """
    Zero-shot: CLIP ko text prompts de do aur wo image classify karega
    Bina kisi fine-tuning ke!
    """
    # Har class ke liye text prompt banao
    # Multiple prompts = better accuracy
    text_prompts = []
    for cn in class_names:
        parts = cn.split('_', 1)
        st = parts[0]
        concern = parts[1] if len(parts) > 1 else cn
        # Multiple templates
        text_prompts.append(f"a photo of {st} skin with {concern}")

    text_tokens = clip.tokenize(text_prompts, truncate=True).to(device)

    all_preds, all_probs, all_true = [], [], []
    clip_model.eval()
    with torch.no_grad():
        # Text features pehle compute karo (ek baar)
        txt_feats = clip_model.encode_text(text_tokens).float()
        txt_feats = F.normalize(txt_feats, dim=-1)

        for batch in tqdm(loader, desc="  Zero-shot inference"):
            imgs = batch['image'].to(device)
            lbls = batch['label']
            img_feats = clip_model.encode_image(imgs).float()
            img_feats = F.normalize(img_feats, dim=-1)

            # Cosine similarity
            sims  = (img_feats @ txt_feats.T) * 100.0
            probs = F.softmax(sims, dim=-1).cpu().numpy()
            preds = sims.argmax(1).cpu().numpy()

            all_probs.extend(probs)
            all_preds.extend(preds)
            all_true.extend(lbls.numpy())

    return np.array(all_true), np.array(all_preds), np.array(all_probs)

zs_true, zs_preds, zs_probs = zero_shot_clip_predict(
    test_loader, CLASS_NAMES, clip_model, device)

print("All baseline models ready!")
```

```python
# ================================================================
# CELL 7: FIXED — CLIP+Adapter predictions collect karo
# Error fix: model ko images + texts dono dene padte hain
# ================================================================

# ---- Helper: PyTorch models ke liye (sirf image use karte hain) ----
def get_predictions_image_only(model_pt, loader):
    """
    ResNet, EfficientNet, ViT, MobileNet ke liye
    Yeh models sirf image lete hain, text nahi
    """
    model_pt.eval()
    all_probs, all_preds, all_true = [], [], []

    with torch.no_grad():
        for batch in tqdm(loader, desc="  Predicting", leave=False):
            imgs  = batch['image'].to(device)
            lbls  = batch['label']
            out   = model_pt(imgs)
            probs = F.softmax(out, dim=-1).cpu().numpy()
            preds = out.argmax(1).cpu().numpy()
            all_probs.extend(probs)
            all_preds.extend(preds)
            all_true.extend(lbls.numpy())

    return np.array(all_true), np.array(all_preds), np.array(all_probs)


# ---- Helper: CLIP+Adapter ke liye (image + text dono chahiye) ----
def get_predictions_clip(model, loader):
    """
    CLIPAdapterModel ke liye — image aur text dono pass karta hai
    """
    model.eval()
    all_probs, all_preds, all_true = [], [], []

    with torch.no_grad():
        for batch in tqdm(loader, desc="  CLIP predicting", leave=False):
            imgs  = batch['image'].to(device)
            txts  = batch['text'].to(device)   # <-- yeh missing tha
            lbls  = batch['label']

            logits, _, _ = model(imgs, txts)   # dono pass karo
            probs = F.softmax(logits, dim=-1).cpu().numpy()
            preds = logits.argmax(1).cpu().numpy()

            all_probs.extend(probs)
            all_preds.extend(preds)
            all_true.extend(lbls.numpy())

    return np.array(all_true), np.array(all_preds), np.array(all_probs)


# ================================================================
# CLIP+Adapter — Standard Classifier Predictions
# ================================================================
print("1. CLIP+Adapter (standard) predictions...")
clip_true_std, clip_preds_std, clip_probs_std = get_predictions_clip(model, test_loader)
print(f"   Done! Accuracy: {accuracy_score(clip_true_std, clip_preds_std)*100:.2f}%")


# ================================================================
# CLIP+Adapter — Few-Shot Predictions
# ================================================================
print("\n2. Building few-shot prototypes...")
model.build_prototypes(train_loader)

clip_preds_fs_list  = []
clip_probs_fs_list  = []
clip_true_fs_list   = []

model.eval()
with torch.no_grad():
    for batch in tqdm(test_loader, desc="   Few-shot inference"):
        imgs = batch['image'].to(device)
        txts = batch['text'].to(device)
        lbls = batch['label']

        pred_labels, sims = model.predict_few_shot(imgs, txts)
        probs = F.softmax(sims, dim=-1).cpu().numpy()
        preds = [model.proto_labels.index(l) for l in pred_labels]

        clip_preds_fs_list.extend(preds)
        clip_probs_fs_list.extend(probs)
        clip_true_fs_list.extend(lbls.numpy())

clip_true_fs  = np.array(clip_true_fs_list)
clip_preds_fs = np.array(clip_preds_fs_list)
clip_probs_fs = np.array(clip_probs_fs_list)

fs_acc = accuracy_score(clip_true_fs, clip_preds_fs) * 100
print(f"   Done! Few-shot Accuracy: {fs_acc:.2f}%")


# ================================================================
# Baseline Models — image only (in models mein text nahi jaata)
# ================================================================
print("\n3. ResNet-50 predictions...")
rn_t, rn_p, rn_pb = get_predictions_image_only(resnet, test_loader)
print(f"   Accuracy: {accuracy_score(rn_t, rn_p)*100:.2f}%")

print("4. EfficientNet predictions...")
en_t, en_p, en_pb = get_predictions_image_only(effnet, test_loader)
print(f"   Accuracy: {accuracy_score(en_t, en_p)*100:.2f}%")

print("5. ViT-B/16 predictions...")
vit_t, vit_p, vit_pb = get_predictions_image_only(vit, test_loader)
print(f"   Accuracy: {accuracy_score(vit_t, vit_p)*100:.2f}%")

print("6. MobileNetV3 predictions...")
mob_t, mob_p, mob_pb = get_predictions_image_only(mobilenet, test_loader)
print(f"   Accuracy: {accuracy_score(mob_t, mob_p)*100:.2f}%")

print("7. SVM predictions (already done in Cell 6)...")
# SVM pehle se Cell 6 mein ho chuka hai

print("8. Zero-Shot CLIP predictions (already done in Cell 6)...")
# Zero-shot bhi Cell 6 mein ho chuka hai

print("\n✅ All predictions collected!")
print("\nSummary:")
print(f"  CLIP+Adapter (std) : {accuracy_score(clip_true_std, clip_preds_std)*100:.2f}%")
print(f"  CLIP+Adapter (FS)  : {fs_acc:.2f}%")
print(f"  ResNet-50          : {accuracy_score(rn_t,  rn_p)*100:.2f}%")
print(f"  EfficientNet-B0    : {accuracy_score(en_t,  en_p)*100:.2f}%")
print(f"  ViT-B/16           : {accuracy_score(vit_t, vit_p)*100:.2f}%")
print(f"  MobileNetV3        : {accuracy_score(mob_t, mob_p)*100:.2f}%")
print(f"  SVM+CLIP feats     : {accuracy_score(test_lbls_svm, svm_preds)*100:.2f}%")
print(f"  Zero-Shot CLIP     : {accuracy_score(zs_true, zs_preds)*100:.2f}%")
```

```python
# ================================================================
# CELL 9: MODEL SAVE + FINAL RECOMMENDER
# ================================================================

# Save
torch.save({
    'img_adapter' : model.img_adapter.state_dict(),
    'txt_adapter' : model.txt_adapter.state_dict(),
    'fusion'      : model.fusion.state_dict(),
    'classifier'  : model.classifier.state_dict(),
    'label2idx'   : train_ds.label2idx,
    'idx2label'   : train_ds.idx2label,
    'num_classes' : NUM_CLASSES,
    'class_names' : CLASS_NAMES,
}, 'skincare_clip_adapter_fyp.pth')
print("Model saved!")


# ---- Product database ----
products_df = pd.DataFrame([
    {"name":"Oil-Free Moisturizer","brand":"CeraVe",
     "skin_type":"oily","concern":"acne","price":2500,"rating":4.5,
     "ingredients":"Niacinamide, Salicylic Acid"},
    {"name":"Pore Minimizing Toner","brand":"Paula's Choice",
     "skin_type":"oily","concern":"pores","price":3200,"rating":4.7,
     "ingredients":"BHA, Niacinamide"},
    {"name":"Rich Hydrating Cream","brand":"Neutrogena",
     "skin_type":"dry","concern":"wrinkles","price":1800,"rating":4.3,
     "ingredients":"Hyaluronic Acid, Ceramides"},
    {"name":"Vitamin C Serum","brand":"The Ordinary",
     "skin_type":"dry","concern":"dark_spots","price":1200,"rating":4.4,
     "ingredients":"15% Vitamin C, Ferulic Acid"},
    {"name":"Calming Face Wash","brand":"La Roche-Posay",
     "skin_type":"sensitive","concern":"redness","price":2800,"rating":4.6,
     "ingredients":"Thermal Water, Ceramides"},
    {"name":"Balancing Moisturizer","brand":"Clinique",
     "skin_type":"combination","concern":"pores","price":3500,"rating":4.4,
     "ingredients":"Niacinamide, Zinc"},
    {"name":"Daily SPF Moisturizer","brand":"EltaMD",
     "skin_type":"normal","concern":"dark_spots","price":4000,"rating":4.8,
     "ingredients":"Vitamin C, SPF 46"},
])


def recommend(image_path, text_query=None, top_k=3):
    """
    Recommendation function:
    image_path  : skin image ka path
    text_query  : optional text (e.g. "my skin is oily and I have acne")
    top_k       : kitne products recommend karein
    """
    img = Image.open(image_path).convert('RGB')
    img_tensor = preprocess(img).unsqueeze(0).to(device)

    if text_query is None:
        text_query = "skin image showing skin condition"
    txt_token = clip.tokenize([text_query], truncate=True).to(device)

    model.eval()
    with torch.no_grad():
        logits, _, _ = model(img_tensor, txt_token)
        probs = F.softmax(logits, dim=-1)
        conf, idx = probs.max(-1)

    pred_label = train_ds.idx2label[idx.item()]
    parts      = pred_label.split('_', 1)
    skin_type  = parts[0]
    concern    = parts[1] if len(parts)>1 else 'general'

    print(f"\n--- Skin Analysis ---")
    print(f"Skin Type : {skin_type.upper()}")
    print(f"Concern   : {concern.upper()}")
    print(f"Confidence: {conf.item()*100:.1f}%")

    match = products_df[
        (products_df.skin_type == skin_type) &
        (products_df.concern   == concern)
    ].sort_values('rating', ascending=False)

    if len(match) == 0:
        match = products_df[products_df.skin_type == skin_type]\
                    .sort_values('rating', ascending=False)

    print(f"\n--- Top {top_k} Recommended Products ---")
    for i, (_, p) in enumerate(match.head(top_k).iterrows(), 1):
        print(f"\n{i}. {p['name']} by {p['brand']}")
        print(f"   Rating     : {p['rating']}/5.0")
        print(f"   Ingredients: {p['ingredients']}")
        print(f"   Price      : Rs. {p['price']:,}")

    return {'skin_type': skin_type, 'concern': concern,
            'confidence': conf.item()}


# Test recommendation
sample_img = test_df.iloc[0]['image_path']
result = recommend(sample_img, "My skin feels very oily and I have breakouts", top_k=3)
```
